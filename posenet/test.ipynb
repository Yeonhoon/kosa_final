- 주기에서 두 점의 기울기 >=0 : 내려감
- 주기에서 두 점의 기울기 < 0 : 올라감

```python
import pandas as pd
df = pd.read_csv('csv/angle_df.csv')
df.head()
```

```python
import seaborn as sns
sns.lineplot(data=df, x='frame', y= 'angles_l')

```

```python
import scipy as sc
import numpy as np
x_new = np.linspace(0, len(df['frame']),10000)
smooting = sc.interpolate.make_interp_spline(df['frame'], df['angles_l'])
y_new= smooting(x_new)
import matplotlib.pyplot as plt
plt.plot(x_new, y_new)
plt.show()
```

```python
import tensorflow as tf
import tensorflow_hub as hub

# Load the input image.
image_path = 'video/SQUAT_2.mp4'
image = tf.io.read_file(image_path)
image = tf.compat.v1.image.decode_jpeg(image)
image = tf.expand_dims(image, axis=0)
# Resize and pad the image to keep the aspect ratio and fit the expected size.
image = tf.cast(tf.image.resize_with_pad(image, 192, 192), dtype=tf.int32)

# Download the model from TF Hub.
model = hub.load("https://tfhub.dev/google/movenet/singlepose/lightning/3")
movenet = model.signatures['serving_default']

# Run model inference.
outputs = movenet(image)
# Output is a [1, 1, 17, 3] tensor.
keypoints = outputs['output_0']
```
